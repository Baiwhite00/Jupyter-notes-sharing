# Oscillation

What is oscilation?


An object oscillaties when it moves back and forth repeatedly between two states which are on either side of some equilibrium position.
Examples: swinging pendulum, alternating current (AC), the atoms of a solid vibrate.


Why Oscillation is important?


Except the physical examples above, oscillations also occurs in dynamic systems in virtually every area of science: for example the beating of the human heart (for circulation), business cycles in economics, predatorâ€“prey population cycles in ecology, and periodic firing of nerve cells in the brain.

## Simple harmonic motion

Simple harmonic motion (s.h.m) is a fundamental type of periodic. The requirements for s.h.m of a mechnical system include:
<br>
<ol>
  <li>a mass that oscillates</li>
  <li>a position where the mass is in equilibrium</li>
  <li>a restoring force that acts to return the mass to its equilibrium position; the restoring force F is directly proportional to the displacement x of the mass from its equilibrium position and is directed towards that point.</li>
</ol>


An canonical question is how to precisely describe the oscillating systems. Many the oscillationg systems can be represented by a displacement-time graph. 

```python
vibrating = Image.open('displacement-time.png')
new_image = vibrating.resize((800, 500))
new_image
```

In this plot, the displacement $x$ varies symmetrically on the either side of the midpoint. The shape is a sine curve and the motion is named as sinusoidal.<br>
<br>
$\color{red}{Amplitude}$ $A$: The maximum displacement from the equilibrium position.<br>
$\color{red}{Period}$ $T$: The time for one complete oscillation.<br>
$\color{red}{Frequency}$ $f$: The number of oscillations per unit time.<br>
By definition, the frequency $T$ is the reciprcal of period $T$ and vice versa:<br>
    <h1><center>$f=\frac{1}{T}$,  $T=\frac{1}{f}$</center></h1>

If two oscillations have the same Amplitude and period, they are not necessarily equivalent. They can have a so called
$\color{red}{\text{phase difference}}$. The term $\color{red}{\text{phase}}$ represents where the oscillating object positioned in the complete cycle of an oscillation.
